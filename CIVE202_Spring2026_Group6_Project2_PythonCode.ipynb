```python
import pandas as pd
import numpy as np
```

```python
cubic_yard_ft3 = 27
unit_weight_water = 62.4
```

**Function Definitions**

```python
#function to calculate water weight
def water_weight_calc(A,B,C,D,E):
    cementious_weight = A+B+C+D
    water_weight = cementious_weight*E
    return water_weight
```

```python
#functions to calculate volumes of cement,fly ash,silica fume, and other scm
def calc_cem_vol(A,J):
    R = A/(J*62.4)
    return R

def calc_fly_ash_vol(fly_ashB,fly_ashK):
    S = fly_ashB/(fly_ashK*62.4)
    return S

def calc_silica_fume_vol(silica_fumeC,silica_fumeL):
    T = silica_fumeC/(silica_fumeL*62.4)
    return T

def calc_scm_vol(scmD,scmM):
    U = scmD/(scmM*62.4)
    return U
```

```python
#functions for air and water portions in concrete
def calc_air_vol(air_contentF):
    V = (air_contentF/100)*27
    return V

def calc_water_vol(water_weight_Q):
    W = water_weight_Q/62.4
    return W
```

```python
#function for total agg volume 
def calc_agg_vol(R,S,T,U,V,W):
    X = 27 - R - S - T - U - V - W
    return X
```

```python
#function for weights of agg components
def calc_fine_agg_weight_Y(G,N,X):
    Y = 62.4 * (G/100) * N * X
    return Y

def calc_coarse_agg_weight_Z(H,O,X):
    Z = 62.4 * (H/100) * O * X
    return Z

def calc_other_agg_weight_AA(W,I,P,X):
    AA = W * (I/100) * P * X
    return AA
```

# Running Mix Design

```python
### user input prompts
# Project metadata
project_number = int(input("Enter project number: "))                   
concrete_class = input("Enter the class of the concrete: ")

# Cementitious material inputs (lb per cubic yard)
cement_A = float(input("Enter weight of cement per yard A: "))
fly_ash_B = float(input("Enter weight of fly ash per yard B: "))
silica_fume_C = float(input("Enter weight of silica fume per yard C: "))
other_scm_D = float(input("Enter weight of SCM per yard D: "))

wc_ratio_E = float(input("Enter target water/cement ratio E: "))
air_content_F = float(input("Enter target % air content F: "))

# Aggregate proportions (%)
percent_fine_agg_G = float(input("Enter target % fine agg G: "))
percent_coarse_agg_H = float(input("Enter target % coarse agg H: "))
percent_other_agg_I  = float(input("Enter target % other agg I: "))

# Specific gravities
sg_cement_J  = float(input("Enter sg of cement J: "))
sg_fly_ash_K  = float(input("Enter sg of fly ash K: "))
sg_silica_fume_L  = float(input("Enter sg of silica fume L: "))
sg_other_scm_M  = float(input("Enter sg of other scm M: "))
sg_fine_agg_N  = float(input("Enter sg of fine agg N: "))
sg_coarse_agg_O  = float(input("Enter sg of coarse agg O: "))
sg_other_agg_P  = float(input("Enter sg of other agg P: "))

#calculations
#water weight
Q = water_weight_calc(cement_A,fly_ash_B,silica_fume_C,other_scm_D,wc_ratio_E)

#component volumes
R = calc_cem_vol(cement_A,sg_cement_J)
S = calc_fly_ash_vol(fly_ash_B,sg_fly_ash_K)
T = calc_silica_fume_vol(silica_fume_C,sg_silica_fume_L)
U = calc_scm_vol(other_scm_D,sg_other_scm_M)

#air & water volumes
V = calc_air_vol(air_content_F)
W = calc_water_vol(Q)

#agg volume
X = calc_agg_vol(R,S,T,U,V,W)

#agg component weights
Y = calc_fine_agg_weight_Y(percent_fine_agg_G,sg_fine_agg_N,X)
Z = calc_coarse_agg_weight_Z(percent_coarse_agg_H,sg_coarse_agg_O,X)
AA = calc_other_agg_weight_AA(W,percent_other_agg_I,sg_other_agg_P,X)

# Output: Final Weight Chart
print("\n----------------------------------------------")
print("     NDOT Concrete Mix Design – Weight Summary")
print("          (1 Cubic Yard of Concrete)")
print("----------------------------------------------")
print(f"Project Number:       {project_number}")
print(f"Class of Concrete:    {concrete_class}")
print("")
print(f"Cement (A):           {cement_A:5.1f} lb")
print(f"Fly Ash (B):          {fly_ash_B:5.1f} lb")
print(f"Silica Fume (C):      {silica_fume_C:5.1f} lb")
print(f"Other SCM (D):        {other_scm_D:5.1f} lb")
print("")
print(f"Fine Aggregate (Y):   {Y:5.0f} lb")
print(f"Coarse Aggregate (Z): {Z:5.0f} lb")
print(f"Other Aggregate (AA): {AA:5.0f} lb")
print("")
print(f"Water (0):            {Q:5.0f} lb")
print("----------------------------------------------")
print("End of Mix Design Summary")
```

# Scenarios

### Scenario 1: 47B-3500
- **Source:**
    - Nebraska Department of Transportation. 2017. Supplemental Specifications, Division 1000. Nebraska Department of Transportation. https://www.nebraskatransportation.org/ref-man/supspecem/SuppSpec-1000.pdf
- **Use case:** Concrete Pavements
- **Inputs:**
    - cement_A = 480
    - fly_ash_B = 98
    - silica_fume_C = 0.0
    - wc_ratio_E = 0.48
    - air_content_F = 6.25
    - percent_fine_agg_G = 70
    - percent_coarse_agg_H =  30


```python
Q = water_weight_calc(480,98,0,0,0.48)
R = calc_cem_vol(480,3.15)
S = calc_fly_ash_vol(98,2.4)
V = calc_air_vol(6.25)
W = calc_water_vol(Q)
X = calc_agg_vol(R, S, 0, 0, V, W)

output_table47B = [["Cement", 480, ""],
    ["Water", water_weight_calc(480, 98, 0, 0, 0.48), ""],   
    ["Fine Aggregate", calc_fine_agg_weight_Y(70, 2.60, X), ""],
    ["Coarse Aggregate", calc_coarse_agg_weight_Z(30, 2.65, X), ""],
    ["Air (volumetric)", 6.25, "percent / ft³"]]                                                                            

headers = ["Component", "Value", "Units"]

df = pd.DataFrame(output_table47B, columns=["Component", "Weight (lb/yd3)", "Notes"])
df["Weight (lb/yd3)"] = df["Weight (lb/yd3)"].round(2)
total = df[df["Component"] != "Air (volumetric)"]["Weight (lb/yd3)"].sum()
df.loc[len(df)] = ["Total", round(total, 2), ""]
df
```
